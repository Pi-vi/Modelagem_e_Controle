‚Üí Em ess√™ncia, a ideia b√°sica de uma transformada √© levar o problema desde o espa√ßo f√≠sico para um novo espa√ßo abstrato, no qual a solu√ß√£o √© obtida de forma conveniente.

‚Üí Uma vez resolvendo o problema nesse espa√ßo, utiliza-se uma transformada inversa para retornar para o espa√ßo f√≠sico.


- Transformada

$X(s) = \mathscr{L}\{x(t)\} = \int_{0}^{\infty} e^{-st}x(t)dt$


- Transformada inversa

$x(t) = \mathscr{L}\{X(s)\} = \dfrac{1}{2\pi i}\int_{a-i\infty}^{a+i\infty} e^{st}X(s)ds$
em que $s = a + ib$

Propriedades da transformada de Laplace:

- $\mathscr{L}\{\alpha_1 x_1 + \alpha_2 x_2\} = \alpha_1 X_1 + \alpha_2 X_2$
- $\mathscr{L}\{\alpha x\} = \alpha X$
- $\mathscr{L}\{\dot{x}\} = sX - x(0)$
- $\mathscr{L}\{\ddot{x} \} = s^2 X - sx(0) - x(0)$

Considere o sistema linear invariante no tempo definido pela seguinte equa√ß√£o diferencial:  

$a_0 \dfrac{d^{n}y}{dt^{n}} + a_1 \dfrac{d^{n-1}y}{dt^{n-1}} + ... + a_{n-1} \dfrac{dy}{dt} + a_n y = b_{0}\dfrac{d^{m}x}{dt^{m}} + b_{0}\dfrac{d^{m}x}{dt^{m}} + b_{1}\dfrac{d^{m-1}x}{dt^{m-1}}+ b_{m-1}\dfrac{dx}{dt} + b_mx \qquad (n\geq m)$  

Onde $y$ √© a sa√≠da e $x$ √© a entrada. A **fun√ß√£o de transfer√™ncia** desse sistema √© a rela√ß√£o entre a transformada de Laplace da sa√≠da e a transformada de Laplace da entrada, quando todas as condi√ß√µes iniciais s√£o zero, ou:  

Fun√ß√£o de Transfer√™ncia = $G(s) = \dfrac{\mathscr{L}[sa\acute{i}da]}{\mathscr{L}[entrada]}$

$G(s) = \dfrac{Y(s)}{X(s)} = \dfrac{b_0 s^m + b_1 s^{m-1} + ... + b_{m-1} s + b_m}{a_0 s^n + a_1s^{n-1} + ... + a_{n-1}s + a_n}$

A **fun√ß√£o de transfer√™ncia** de um sistema representado por uma equa√ß√£o diferencial linear invariante no tempo √© definida como a rela√ß√£o entre a transformada de Laplace da sa√≠da (fun√ß√£o de resposta) e a transformada de Laplace da entrada (fun√ß√£o de excita√ß√£o), admitindo-se todas as condi√ß√µes iniciais nulas. 


‚Üí Uma vez determinada, a fun√ß√£o de transfer√™ncia fornece uma descri√ß√£o das caracter√≠sticas din√¢micas do sistema, independentemente de sua descri√ß√£o f√≠sica.

‚Üí A aplicabilidade do conceito de fun√ß√£o de transfer√™ncia √© limitada a sistemas de equa√ß√µes diferenciais lineares invariantes no tempo. 


Para um sistema linear, invariante no tempo, a fun√ß√£o de transfer√™ncia $G(s)$ √©:

$G(s) = \dfrac{Y(s)}{X(s)}$

Onde: 
$X(s)$ ‚Üí √© a transformada de Laplace da entrada (fun√ß√£o de excita√ß√£o)
$Y(s)$ ‚Üí √© a transformada de Laplace da sa√≠da do sistema (fun√ß√£o de resposta)

‚Üí Fun√ß√£o impulso unit√°rio

- A resposta a um impulso pode ser entendida como uma condi√ß√£o inicial do sistema, uma vez que sua aplica√ß√£o ocorre em um instante de tempo determinado e, depois, ela cessa. 

‚Üí A transformada de Laplace da fun√ß√£o impulso unit√°rio √© igual a unidade. 

A transformada de Laplace da fun√ß√£o da sa√≠da do sistema √©:

$Y(s) = G(s)$

A resposta impulsiva do sistema √© ent√£o a transformada inversa da fun√ß√£o de transfer√™ncia.

$\mathscr{L}^{-1}[G(s)] = g(t)$

A fun√ß√£o de resposta impulsiva $g(t)$ √©, portanto, a resposta de um sistema linear invariante no tempo a um impulso unit√°rio de entrara, quando as condi√ß√µes iniciais do sistema s√£o nulas. 

√â poss√≠vel obter informa√ß√µes completas sobre as caracter√≠sticas din√¢micas de um sistema, por meio da excita√ß√£o por um impulso de entrada e medindo a resposta.

## Polos e Zeros de Tranformadas de Laplace

Geralmente, uma transformada de Laplace pode ser expressa usando o formato:

$F(s) = \dfrac{a(s+z_1)(s+z_2)...(s+z_m)}{(s+p_1)(s+p_2)...(s+p_n)}$

Onde $a$ √© uma constante.

Os valores de $s = -z_1$, $s= -z_2$, ..., $s=-z_m$ que fazem com que $F(s) = 0$ s√£o denominados **zeros** da fun√ß√£o de transfer√™ncia.

Os valores de $s = -p_1$, $s = -p_2$, ..., $s=-p_n$ que fazem com que $F(s) = \infty$ s√£o denominados **polos** de $F(s)$.

## Transformada Inversa de Laplace

A resposta do sistema no dom√≠nio de Laplace √© expressa por uma raz√£o de polin√¥mios:

$G(s) = \dfrac{Y(s)}{X(s)} = \dfrac{b_0 s^m + b_1 s^{m-1} + ... + b_{m-1} s + b_m}{a_0 s^n + a_1s^{n-1} + ... + a_{n-1}s + a_n}$

Como o objetivo final √© efetuar a transforma√ß√£o inversa, √© util utilizar a expans√£o em fra√ß√µes parciais.

Para isso, √© conveniente expressar o denominador D(s) a partir das suas ra√≠zes, conhecidas como polos do sistema. Assim: 

$X(s) = \dfrac{N(s)}{(s+p_1)(s+p_2)...(s+p_3)}$

$X(s) = \dfrac{C_1}{s+p_1} + \dfrac{C_2}{s+p_2} + ... + \dfrac{C_j}{s+p_n}$

$x(t) = C_1 e^{-p_1 t} + C_2 e ^{-p_2 t} + ... + C_n  e^{p_n t}$

## Resposta a uma Fun√ß√£o Impulso Unit√°rio

A transformada de Laplace da fun√ß√£o Impulso Unit√°rio (delta de Dirac) √© igual a 1.

**Exemplo 2:**

Avalie a resposta de um oscilador linear (sistema massa-mola-amortecedor) submetido a um impulso utilizando a transformada de Laplace.

Dados:
* $m = 2kg$
* $\zeta = 0.15$
* $\omega_n = 30$

---

````python
s,t = symbols('s t')
````

````python
m = 2
zeta = 0.15
omega_n = 30
````

````python
G = (1/m)/(s**2 + 2*zeta*omega_n*s + omega_n**2)
G
````

````python
G.apart(s)
````

````python
x = inverse_laplace_transform(G,s,t)
x
````

````python
Heaviside(t) # Fun√ß√£o degrau unit√°rio
````

````python
plot(Heaviside(t))
````

````python
DiracDelta(t)
````

````python
x.subs(Heaviside(t),1)
````

````python
plot(x,(t,0,2))
````

````python
num = [1/m]
den = [1,2*zeta*omega_n,omega_n**2]

G = tf(num,den)
G
````

````python
from numpy import linspace
````

````python
tn = linspace(0,2,1000)
t1,y1 = impulse_response(G,tn)
````

````python
import matplotlib.pyplot as plt
````

````python
plt.figure(dpi = 300)
plt.plot(t1,y1)
plt.xlabel('tempo[s]')
plt.ylabel('Amplitude')
plt.title('Resposta do sistema massa-mola-amortecedor a uma fun√ß√£o impulso')
plt.grid(True)
````
